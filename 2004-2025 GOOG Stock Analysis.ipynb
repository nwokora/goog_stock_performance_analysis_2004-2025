## **Libraries import**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## **Read csv file**

```python
cols = ['date', 'open', 'high', 'low', 'close', 'adj_close', 'volume']
df = pd.read_csv('GOOG_2004-08-19_2025-08-20.csv', skiprows=7, names=cols)
df.head()
```

```python
print(len(df))
```

## **Datetime standardisation**

```python
df["date"] = pd.to_datetime(df["date"])
```

## **Date column indexing**

```python
df = df.set_index("date").sort_index()
print(df)
```

## **Missing value check**

```python
df.isna().sum()
```

## **Duplicates check**

```python
df.index.duplicated().sum()
```

## **Daily price range**

```python
df["range"] = df["high"] - df["low"]
print(df)
```

```python
df["Year"] = df.index.year
print(df)
```

```python
df["return"] = df["close"].pct_change()
```

```python
df
```

```python
pd.set_option('expand_frame_repr', False)
print(df.head(20))
```

```python
df.to_csv('2004-2025_goog_cleaned_dataset.csv')
```

## **Total Trading Days**

```python
days = np.busday_count('2004-08-19','2025-08-19')
print(f'Trading days: {days}')
```

## **Win Rate Percentage (close > open)**

```python
df['Win'] = df['close'] > df['open']
win_rate = df['Win'].mean() * 100

print(f"Close > Open: {win_rate:.1f}%")
```

## **Daily Volatility**

```python
daily_volatility = df['return'].std() * 100

print(f"Daily Volatility: {daily_volatility:.2f}%")
```

## **Q1: Total Trading Volume by Year**

```python
yearly_volume = df.groupby(df.index.year)['volume'].sum()

total_volume = df['volume'].sum()
print(f"Total-volume: {total_volume:,} shares")

max_year = yearly_volume.idxmax()
min_year = yearly_volume.idxmin()
max_vol = yearly_volume.loc[max_year]
min_vol = yearly_volume.loc[min_year]

print(f"Highest‑volume year: {max_year} with {max_vol:,} shares")
print(f"Lowest‑volume year: {min_year} with {min_vol:,} shares")

plt.figure(figsize=(8, 3))
plt.bar(yearly_volume.index, yearly_volume.values)
plt.title("Total Trading Volume by Year")
plt.xlabel("Year")
plt.ylabel("Total volume")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## **Q4: Average Daily Return by Year**

```python
avg_daily_ret_by_year = df.groupby(df.index.year)['return'].mean()

max_year = avg_daily_ret_by_year.idxmax()
min_year = avg_daily_ret_by_year.idxmin()
max_ret = avg_daily_ret_by_year.loc[max_year]
min_ret = avg_daily_ret_by_year.loc[min_year]

print(f"Highest avg daily return year: {max_year} = {max_ret:+.4f}")
print(f"Lowest avg daily return year: {min_year} = {min_ret:+.4f}")

plt.figure(figsize=(8, 3.5))
plt.bar(avg_daily_ret_by_year.index, avg_daily_ret_by_year.values)
plt.title("Average Daily Return by Year")
plt.xlabel("Year")
plt.ylabel("Average daily return")
plt.grid(axis="y", alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **Q5: Average Daily Return by Month**

```python
monthly_seasonality = df.groupby(df.index.month)["return"].mean()

monthly_seasonality.index = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                             "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

max_month = monthly_seasonality.idxmax()
min_month = monthly_seasonality.idxmin()
max_ret = monthly_seasonality.loc[max_month]
min_ret = monthly_seasonality.loc[min_month]

print(f"Highest‑return month: {max_month} = {max_ret:+.4f}")
print(f"Lowest‑return month:  {min_month} = {min_ret:+.4f}")

plt.figure(figsize=(8, 3.1))
plt.bar(monthly_seasonality.index, monthly_seasonality.values)
plt.xlabel("Month")
plt.ylabel("Average daily return")
plt.title("Average Daily Return by Month")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.show()
```
